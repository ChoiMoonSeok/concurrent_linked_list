## import libs

```python
import matplotlib.pyplot as plt
```

## values

```python
avg_spinlock_coarse_grained_add = [125438, 1474126, 3738624, 12551165, 28268722, 40759941]
avg_spinlock_coarse_grained_search = [126684, 1477123, 3740093, 9645129, 22083905, 31261239]

avg_spinlock_hand_over_hand_add = [331924, 4644774, 8702925, 9658266, 10352693, 9983535]
avg_spinlock_hand_over_hand_search = [338554, 4249573, 9813694, 10306421, 10240545, 10674572]
```

```python
avg_mutex_coarse_grained_add = [125683, 306562, 480940, 878183, 1008858, 1027683]
avg_mutex_coarse_grained_search = [125691, 443728, 565298, 1354096, 1332423, 1661853]

avg_mutex_hand_over_hand_add = [674267, 5162058, 11649237, 7404249, 11566500, 11728626]
avg_mutex_hand_over_hand_search = [689991, 5762103, 12329818, 9193132, 10032830, 11868520]
```

```python
plt.plot([1, 10, 20, 30, 40, 50], avg_spinlock_coarse_grained_add)
plt.plot([1, 10, 20, 30, 40, 50], avg_mutex_coarse_grained_add)
plt.plot([1, 10, 20, 30, 40, 50], avg_spinlock_hand_over_hand_add)
plt.plot([1, 10, 20, 30, 40, 50], avg_mutex_hand_over_hand_add)
plt.legend(['spin lock coarse-grained', 'spin lock hand-over-hand', 'mutex lock coarse-grained', 'mutex lock hand-over-hand'])
plt.xlabel('Number of threads')
plt.ylabel('Average add time')
plt.title('Comparision of average add time by locking algorithm')
plt.show()
```

```python
plt.plot([1, 10, 20, 30, 40, 50], avg_spinlock_coarse_grained_search)
plt.plot([1, 10, 20, 30, 40, 50], avg_mutex_coarse_grained_search)
plt.plot([1, 10, 20, 30, 40, 50], avg_spinlock_hand_over_hand_search)
plt.plot([1, 10, 20, 30, 40, 50], avg_mutex_hand_over_hand_search)
plt.legend(['spin lock coarse-grained', 'spin lock hand-over-hand', 'mutex lock coarse-grained', 'mutex lock hand-over-hand'])
plt.xlabel('Number of threads')
plt.ylabel('Average add time')
plt.title('Comparision of average search time by locking algorithm')
plt.show()
```
